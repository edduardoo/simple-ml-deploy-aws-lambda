```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

area = np.array([1150, 2220, 2555, 3250, 3590, 4689, 5894, 7812, 9879, 10222]).reshape(-1, 1)
price = np.array([125000, 215000,195000, 350000,295000,398000, 598000, 745000, 995000, 1120000]).reshape(-1, 1)

lin_reg = linear_model.LinearRegression()
lin_reg.fit(area, price)

# Make predictions using the testing set
y_pred = lin_reg.predict(area)

# The coefficients
print("Coefficients: \n", lin_reg.coef_)
# The mean squared error
print("Mean squared error: %.2f" % lin_reg.score(area, price))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(y_pred, price))

# Plot outputs
plt.scatter(area, price, color="black")
plt.plot(area, y_pred, color="blue", linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```
